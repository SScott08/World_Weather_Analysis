```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
import random
import numpy as np
import timeit
import matplotlib.pyplot as plt
import json
import csv
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
# Import the requests library.
import requests
from config import weather_api_key 
```

```python
#3 Create a new set of 2,000 random latitudes and longitudes..
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python

coordinates = list(lat_lngs)
coordinates
```

```python
cities = []
```

```python
#4 Get the nearest city using the citipy module.
for coordinate in coordinates:
    city=citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in cities:
        cities.append(citipy.nearest_city(coordinate[0], coordinate[1]).city_name)
        print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
#5 Perform an API call with the OpenWeatherMap.
latitude =[]
longitude = []
city_list = []
temperature = []
humidity = []
cloudiness = []
wind = []
description = []
country =[]

for city in cities:
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key + "&q=" + city
    try:   
        weather = requests.get (url).json()
        #print (weather)
        #break
        print (city)
        latitude.append(weather["coord"]["lat"])
        longitude.append(weather["coord"]["lon"])
        temperature.append(weather["main"]["temp_max"])
        humidity.append(weather["main"]["humidity"])
        cloudiness.append(weather["clouds"]["all"])
        wind.append(weather["wind"]["speed"])
        description.append(weather["weather"][0]["description"])
        country.append(weather["sys"]["country"])
        city_list.append(city)
     
    except:
        print(city + " not found")
weather_df = pd.DataFrame({
    
    "city": city_list,
    "latitude":latitude, 
    "longitude":longitude,
    "temperature":temperature,
    "humidity":humidity,
    "cloudiness":cloudiness,
    "wind":wind,
    "description":description,
    "country":country
    
})    
weather_df.head()
```

```python
weather_df.to_csv("WeatherPy_Database.csv")
```
